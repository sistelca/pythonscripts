```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
#from PIL import Image
```

```python
image = cv.imread('./Pictures/prueba/Satellite-Earth.jpg')

plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))

plt.show()
```

```python
# RGB a gris
img_gris = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

status = cv.imwrite('./Pictures/prueba/Earth_gris.jpg', img_gris)

print("Imagen salvada gris : ", status)

image_gris = cv.imread('./Pictures/prueba/Earth_gris.jpg')

plt.imshow(image_gris)
plt.show()
```

```python
# convertir a HSV

img_gris = cv.imread('./Pictures/prueba/Satellite-Earth.jpg', cv.IMREAD_GRAYSCALE)


# convertir a HSV
img_bn = cv.adaptiveThreshold(img_gris,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)

status = cv.imwrite('./Pictures/prueba/Earth_bn.jpg', img_bn)

print("Imagen salvada bn : ", status)

image_gris = cv.imread('./Pictures/prueba/Earth_bn.jpg')

plt.imshow(image_gris)
plt.show()
```

```python
# de bgr a bgr
im_bgr = cv.imread('./Pictures/prueba/Satellite-Earth.jpg')
fixed = cv.cvtColor(im_bgr, cv.COLOR_RGB2BGR)

status = cv.imwrite('./Pictures/prueba/Earth_bGR.jpg', fixed)

print("Imagen salvada bgr : ", status)

img = cv.imread('./Pictures/prueba/Earth_bGR.jpg')

plt.imshow(img)
plt.show()
```

```python
img = cv.imread('./Pictures/prueba/Satellite-Earth.jpg')
rect = cv.rectangle(img,(84,400),(670,528),(0,0,255),3)
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img,'hola mundo',(90,500), font, 3,(255,255,255),2,cv.LINE_AA)

cv.imwrite('./Pictures/prueba/Earth_fig.jpg', img)

plt.imshow(img)
plt.show()
```

```python
img = cv.imread('./Pictures/prueba/Satellite-Earth.jpg')
mask = cv.imread('./Pictures/prueba/Earth_mask.jpg')
res = cv.bitwise_and(img,mask)
```

```python
cv.imwrite('./Pictures/prueba/Earth_masked.jpg', res)

plt.imshow(res)
plt.show()
```
