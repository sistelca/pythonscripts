```python
import pandas as pd
import os
import re
```

```python
archivos = [file for file in os.listdir('datos/alphasense') if '.csv' in file]
for arch in archivos:
    print(arch[:-4])
```

```python
def crea_py(file):
    file_orig = file
    file = file.replace(' ', '')
    pyfile = [archi for archi in os.listdir('datos/alphasense') if '.py' in archi]
    tab = '    '
    salto = repr('\n')
    
    if not file[:-4]+'.py' in pyfile:
        with open(os.path.join('datos/alphasense', file[:-4]+'.py'), 'w') as pf:
            pf.write(f"""import pandas as pd"""+'\n')
            pf.write(f"""import os"""+'\n')
            pf.write(f"""import re"""+'\n')
            pf.write('\n')
            pf.write(f"""# {file[:-4]} """+'\n')
            pf.write('\n')
            
            pf.write(f"""def lea_rchivo(file):"""+'\n')
            pf.write(f"""{tab}with open(file, 'r') as file:"""+'\n')
            pf.write(f"""{tab}{tab}nombre = ''"""+'\n')
            pf.write(f"""{tab}{tab}columnas = []"""+'\n')
            pf.write(f"""{tab}{tab}datos = []"""+'\n')
            pf.write(f"""{tab}{tab}i = 0"""+'\n')
            pf.write(f"""{tab}{tab}for linea in file:"""+'\n')
            pf.write(f"""{tab}{tab}{tab}if i == 0:"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}nombre = linea.split(';')[0]"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}nombre = ('').join(re.findall('([A-Z]*[ ]*)', nombre))"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}nombre = nombre.replace(' ', '_')"""+'\n')
            pf.write(f"""{tab}{tab}{tab}elif i == 1:"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}columnas = linea.replace({salto}, '').split(';')"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}columnas = [('').join(re.findall('([A-Z]*[a-z]*[0-9]*[ ]*)', col)) for col in columnas]"""+'\n')
            pf.write(f"""{tab}{tab}{tab}else:"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}dato = linea.replace({salto}, '').split(';')"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}dato[0] = dato[0].replace('"', '')"""+'\n')
            pf.write(f"""{tab}{tab}{tab}{tab}datos.append(dato)"""+'\n')
            pf.write(f"""{tab}{tab}{tab}i += 1"""+'\n')
            pf.write(f"""{tab}{tab}return (nombre, columnas, datos)"""+'\n')
            pf.write('\n')
            pf.write(f"""def crear_df(file):"""+'\n')
            pf.write(f"""{tab}nombre, columnas, datos = lea_rchivo(file)"""+'\n')
            pf.write(f"""{tab}df_name = nombre"""+'\n')
            pf.write(f"""{tab}df_cols = columnas"""+'\n')
            pf.write(f"""{tab}data = datos"""+'\n')
            pf.write('\n')
            pf.write(f"""{tab}df = pd.DataFrame(data, columns=df_cols)"""+'\n')
            pf.write(f"""{tab}df[df_cols[0]] = pd.to_datetime(df[df_cols[0]], dayfirst=True)"""+'\n')
            pf.write('\n')
            pf.write(f"""{tab}for i in range(1, len(df_cols[1:])+1):"""+'\n')
            pf.write(f"""{tab}{tab}df[df_cols[i]] = pd.to_numeric(df[df_cols[i]], errors='coerce')"""+'\n')
            pf.write('\n')
            pf.write(f"""{tab}df['ubicacion'] = df_name"""+'\n')
            pf.write('\n')
            pf.write(f"""{tab}return df"""+'\n')
            pf.write('\n')
            pf.write(f"""df_{file[:-4]} = crear_df('{file_orig}')"""+'\n')
            pf.write(f"""print(df_{file[:-4]}.head())"""+'\n')
            
def lea_rchivo(file):
    with open(os.path.join('datos/alphasense', file), 'r') as file:

        nombre = ''
        columnas = []
        datos = []
        
        i = 0
        for linea in file:
            if i == 0:
                nombre = linea.split(';')[0]
                nombre = ('').join(re.findall('([A-Z]*[ ]*)', nombre))
                nombre = nombre.replace(' ', '_')
            elif i == 1:
                columnas = linea.replace('\n', '').split(';')
                columnas = [('').join(re.findall('([A-Z]*[a-z]*[0-9]*[ ]*)', col)) for col in columnas]
            else:
                dato = linea.replace('\n', '').split(';')
                dato[0] = dato[0].replace('"', '')
                datos.append(dato)
        
            i += 1

        return (nombre, columnas, datos)
```

```python
def crear_df(file):

    nombre, columnas, datos = lea_rchivo(file)
    df_name = nombre
    df_cols = columnas
    data = datos
    
    df = pd.DataFrame(data, columns=df_cols)
    df[df_cols[0]] = pd.to_datetime(df[df_cols[0]], dayfirst=True)
    
    for i in range(1, len(df_cols[1:])+1):
        df[df_cols[i]] = pd.to_numeric(df[df_cols[i]], errors='coerce')
    
    df['ubicacion'] = df_name
    
    return df
```

```python
df_master = pd.DataFrame()

for archivo in archivos:
    print(archivo)
    #df = crear_df(archivo)
    #df_master = pd.concat([df_master, df], ignore_index=True)
    crea_py(archivo)
```
