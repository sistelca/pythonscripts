```python
import re
```

```python
with open('Respuesta Davivienda (1).txt', 'r') as file:
    
    contenido = file.readlines()
    
    listo = []
    
    for linea in contenido:
        
        partes = str(linea).replace('\n', '').split('   ')
        partes = [x for x in partes if len(x.strip()) > 0]
        numero1 = ('').join(re.findall('([0-9]*)', partes[1][:25]))
        nombre = ('').join(re.findall('([A-Z]*[ ]*)', partes[1]))
        nombre = nombre.strip()
        numero2 = ('').join(re.findall('([0-9]*)', partes[2].split(' ')[0]))
        resto_string = linea[linea.find(nombre)+len(nombre):]

        cosa = [x for x in resto_string.split(' ') if len(x)> 0]
                
        tcos3 = ' '.join(cosa[2:])
        
        cosa[3] = [x for x in ['Rechazado', 'Aplicado'] if tcos3.find(x) > 0][0]
        print(cosa[4])
        cosa[2] = tcos3[:tcos3.find(cosa[3])]
        
        soluc = [partes[0], numero1, nombre, cosa[0], cosa[1], cosa[2], cosa[3]]        

        listo.append(soluc)
        print(soluc)


```
